```python
import pandas as pd
import sqlalchemy
import pprint
import psycopg2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sqlalchemy import create_engine
```

```python
# connect to local database
# https://docs.sqlalchemy.org/en/14/core/engines.html#postgresql
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost:5432/sql_challenge')
connection = engine.connect()
```

```python
# Test connection on first table in ERD
departments = pd.read_sql('SELECT * FROM departments', connection)
departments
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
# Based on ERD, should only require salaries table
salaries = pd.read_sql('SELECT * FROM salaries', connection)
salaries
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html
def histogram():
    plt.figure(figsize = (10,6))
    plt.axvline(salaries['salary'].mean(), color='k', linestyle='dashed', linewidth=1, label="Salary Mean")
    plt.xlabel("Salary", fontsize = 12)
    plt.ylabel("Number of Employees", fontsize = 12)
    plt.legend(loc = 0)
    plt.grid(zorder = 0)
    plt.title("Common Salary Ranges", fontweight = 'bold', fontsize = 12)
    plt.hist(salaries['salary'], 20, density=False, label="Salary")
    plt.savefig("Common Salary Ranges.png")
```

```python
histogram()
print(f"The most common salary appears to be $40k. The mean salary in the data is ${'${:0,.0f}'.format(salaries['salary'].mean())}.\
\nHowever, one of the limitations of the data is that the employees dataset only considers their hire date.\
\nIt is possible (and could be confirmed through further analysis) that the bulk of the $40k salaried employees\
\nwere hired at an earlier time, and that the salary graduation towards $100k is a result of income growth over time.")
```

```python
# Create a bar chart of average salary by title.
# import titles and employees as df to merge
titles = pd.read_sql('SELECT * FROM titles', connection)
employees = pd.read_sql('SELECT * FROM employees', connection)
```

```python
titles
```

```python
employees.head()
```

```python
merge_1 = employees.merge(titles, how = 'inner', left_on = 'emp_title_id', right_on = 'title_id')
merge_1.head()
```

```python
merge_2 = pd.merge(merge_1, salaries, on = 'emp_no')
merge_2.head()
```

```python
title_salary = merge_2[["title","salary"]]
avg_salary_title = title_salary.groupby('title').mean().reset_index()
avg_salary_title
```

```python
def bar():
    plt.figure(figsize = (10,6))
    plt.xlabel("Position Title", fontsize = 12)
    plt.ylabel("Average Salary", fontsize = 12)
    plt.grid(zorder = 0)
    plt.title("Average Salaries by Position Title", fontweight = 'bold', fontsize = 12)
    plt.bar(avg_salary_title['title'], avg_salary_title['salary'], zorder = 3)
    plt.savefig("Average Salaries by Position Title.png")
```

```python
bar()
print(f"This plot shows the average salaries per position title. It would appear that Senior Staff and Staff positions\
\npaid the most on average, with Senior Engineers paid the least.")
```
